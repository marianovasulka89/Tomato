```python
her_AE_CE = pd.DataFrame(columns=['enzima' , 'A_LxC' , 'D_LxC' , 'D/A_LxC' ,'A_LA0147xP' , 'D_LA0147xP',
'D/A_LA0147xP', 'A_FxLA1420' , 'D_FxLA1420', 'D/A_FxLA1420'])

her_AE_OR = pd.DataFrame(columns=['enzima' , 'A_LxC' , 'D_LxC' ,'D/A_LxC' , 'A_LA0147xP' , 'D_LA0147xP',
'D/A_LA0147xP','A_FxLA1420' , 'D_FxLA1420' , 'D/A_FxLA1420'])

her_met_CE = pd.DataFrame(columns=['metabolito' , 'A_LxC' , 'D_LxC' ,'D/A_LxC' , 'A_LA0147xP' , 'D_LA0147xP', 
'D/A_LA0147xP','A_FxLA1420' , 'D_FxLA1420' , 'D/A_FxLA1420'])

her_met_OR = pd.DataFrame(columns=['metabolito' , 'A_LxC' , 'D_LxC' ,'D/A_LxC' , 'A_LA0147xP' , 'D_LA0147xP', 
'D/A_LA0147xP', 'A_FxLA1420' , 'D_FxLA1420' , 'D/A_FxLA1420'])

her_met_OR.columns
```

```python
her_AE_CE['enzima'] = AE['enzima']

her_AE_OR['enzima'] = AE['enzima']

her_met_CE['metabolito'] = met['metabolito']

her_met_OR['metabolito'] = met['metabolito']

her_met_OR['metabolito']
```

```python
AE_OR.columns
```

```python
#[A = (P1-P2)/2]

#[D = H-((P1+P2)/2)]

#Levovil (P1) x Cervil (P2), LA0147 (P1) x Plovdiv (P2) and Ferum (P1) x LA1420 (P2)
#aca aggrego el A_geno.abs() para hebitar incordios con los signos
#para LxC , AE

her_AE_CE['A_LxC'] = ((AE_CE['LevovilMeanCE'] - AE_CE['CervilMeanCE'])/2)

her_AE_CE['A_LxC'] = her_AE_CE['A_LxC'].abs()

her_AE_CE['D_LxC'] = (AE_CE['LevovilxCervilMeanCE'] - ((AE_CE['LevovilMeanCE'] + AE_CE['CervilMeanCE'])/2))

#para LxC , met

her_met_CE['A_LxC'] = ((met_CE['LevovilMeanCE'] - met_CE['CervilMeanCE'])/2)

her_met_CE['A_LxC'] = her_met_CE['A_LxC'].abs()

her_met_CE['D_LxC'] = (met_CE['LevovilxCervilMeanCE'] - ((met_CE['LevovilMeanCE'] + met_CE['CervilMeanCE'])/2))

#LA0147xPlovdiv

#para AE

her_AE_CE['A_LA0147xP'] = ((AE_CE['LA0147MeanCE'] - AE_CE['PlovdivMeanCE'])/2)

her_AE_CE['A_LA0147xP'] = her_AE_CE['A_LA0147xP'].abs()

her_AE_CE['D_LA0147xP'] = (AE_CE['LA0147xPlovdivMeanCE'] - ((AE_CE['LA0147MeanCE'] + AE_CE['PlovdivMeanCE'])/2))

#para met

her_met_CE['A_LA0147xP'] = ((met_CE['LA0147MeanCE'] - met_CE['PlovdivMeanCE'])/2)

her_met_CE['A_LA0147xP'] = her_met_CE['A_LA0147xP'].abs()

her_met_CE['D_LA0147xP'] = (met_CE['LA0147xPlovdivMeanCE'] - ((met_CE['LA0147MeanCE'] + met_CE['PlovdivMeanCE'])/2))

#FerumxLA1420

#para AE

her_AE_CE['A_FxLA1420'] = ((AE_CE['FerumMeanCE'] - AE_CE['LA1420MeanCE'])/2)

her_AE_CE['A_FxLA1420'] = her_AE_CE['A_FxLA1420'].abs()

her_AE_CE['D_FxLA1420'] = (AE_CE['FerumxLA1420MeanCE'] - ((AE_CE['FerumMeanCE'] + AE_CE['LA1420MeanCE'])/2))

#para met

her_met_CE['A_FxLA1420'] = ((met_CE['FerumMeanCE'] - met_CE['LA1420MeanCE'])/2)

her_met_CE['A_FxLA1420'] = her_met_CE['A_FxLA1420'].abs()

her_met_CE['D_FxLA1420'] = (met_CE['FerumxLA1420MeanCE'] - ((met_CE['FerumMeanCE'] + met_CE['LA1420MeanCE'])/2))

#####para OR#####

#para LxC , AE

her_AE_OR['A_LxC'] = ((AE_OR['LevovilMeanOR'] - AE_OR['CervilMeanOR'])/2)

her_AE_OR['A_LxC'] = her_AE_OR['A_LxC'].abs()

her_AE_OR['D_LxC'] = (AE_OR['LevovilxCervilMeanOR'] - ((AE_OR['LevovilMeanOR'] + AE_OR['CervilMeanOR'])/2))

#para LxC , met

her_met_OR['A_LxC'] = ((met_OR['LevovilMeanOR'] - met_OR['CervilMeanOR'])/2)

her_met_OR['A_LxC'] = her_met_OR['A_LxC'].abs()

her_met_OR['D_LxC'] = (met_OR['LevovilxCervilMeanOR'] - ((met_OR['LevovilMeanOR'] + met_OR['CervilMeanOR'])/2))

#LA0147xPlovdiv

#para AE

her_AE_OR['A_LA0147xP'] = ((AE_OR['LA0147MeanOR'] - AE_OR['PlovdivMeanOR'])/2)

her_AE_OR['A_LA0147xP'] = her_AE_OR['A_LA0147xP'].abs()

her_AE_OR['D_LA0147xP'] = (AE_OR['LA0147xPlovdivMeanOR'] - ((AE_OR['LA0147MeanOR'] + AE_OR['PlovdivMeanOR'])/2))

#para met

her_met_OR['A_LA0147xP'] = ((met_OR['LA0147MeanOR'] - met_OR['PlovdivMeanOR'])/2)

her_met_OR['A_LA0147xP'] = her_met_OR['A_LA0147xP'].abs()

her_met_OR['D_LA0147xP'] = (met_OR['LA0147xPlovdivMeanOR'] - ((met_OR['LA0147MeanOR'] + met_OR['PlovdivMeanOR'])/2))

#FerumxLA1420

#para AE

her_AE_OR['A_FxLA1420'] = ((AE_OR['FerumMeanOR'] - AE_OR['LA1420MeanOR'])/2)

her_AE_OR['A_FxLA1420'] = her_AE_OR['A_FxLA1420'].abs()

her_AE_OR['D_FxLA1420'] = (AE_OR['FerumxLA1420MeanOR'] - ((AE_OR['FerumMeanOR'] + AE_OR['LA1420MeanOR'])/2))

#para met

her_met_OR['A_FxLA1420'] = ((met_OR['FerumMeanOR'] - met_OR['LA1420MeanOR'])/2)

her_met_OR['A_FxLA1420'] = her_met_OR['A_FxLA1420'].abs()

her_met_OR['D_FxLA1420'] = (met_OR['FerumxLA1420MeanOR'] - ((met_OR['FerumMeanOR'] + met_OR['LA1420MeanOR'])/2))

#siguen los ratios

#LxC

her_AE_CE['D/A_LxC'] = (her_AE_CE['D_LxC']/her_AE_CE['A_LxC'])

her_met_CE['D/A_LxC'] = (her_met_CE['D_LxC']/her_met_CE['A_LxC'])

her_AE_OR['D/A_LxC'] = (her_AE_OR['D_LxC']/her_AE_OR['A_LxC'])

her_met_OR['D/A_LxC'] = (her_met_OR['D_LxC']/her_met_OR['A_LxC'])

#LA0147xP

her_AE_CE['D/A_LA0147xP'] = (her_AE_CE['D_LA0147xP']/her_AE_CE['A_LA0147xP'])

her_met_CE['D/A_LA0147xP'] = (her_met_CE['D_LA0147xP']/her_met_CE['A_LA0147xP'])

her_AE_OR['D/A_LA0147xP'] = (her_AE_OR['D_LA0147xP']/her_AE_OR['A_LA0147xP'])

her_met_OR['D/A_LA0147xP'] = (her_met_OR['D_LA0147xP']/her_met_OR['A_LA0147xP'])

#FxLA1420

her_AE_CE['D/A_FxLA1420'] = (her_AE_CE['D_FxLA1420']/her_AE_CE['A_FxLA1420'])

her_met_CE['D/A_FxLA1420'] = (her_met_CE['D_FxLA1420']/her_met_CE['A_FxLA1420'])

her_AE_OR['D/A_FxLA1420'] = (her_AE_OR['D_FxLA1420']/her_AE_OR['A_FxLA1420'])

her_met_OR['D/A_FxLA1420'] = (her_met_OR['D_FxLA1420']/her_met_OR['A_FxLA1420'])

her_met_OR.head(50)
```
