```python
#de analizar la situacion, conviene reemplazar los nan de las col de p-val por 1 , y los de coef por 0
#ya que al variar su num entre conjunto de genotipos(parentales, o par+hib) podria perder casos al ignorarlos
reemplazo_nan = {'Coef_todos_CE' : 0 , 'Coef_parentales_CE' : 0 , 'Coef_todos_OR' : 0 , 'Coef_parentales_OR' : 0 ,
'PVal_todos_CE' : 1 , 'PVal_parentales_CE' : 1 , 'PVal_todos_OR' : 1 , 'PVal_parentales_OR' : 1}
corr = corr.fillna(value=reemplazo_nan)
```

```python
corrCE = corr.loc[:,['Unicidad', 'Locus', 'Coef_todos_CE', 'PVal_todos_CE',
       'Coef_parentales_CE', 'PVal_parentales_CE']]

corrOR = corr.loc[:,['Unicidad', 'Locus', 'Coef_todos_OR','PVal_todos_OR', 'Coef_parentales_OR', 
'PVal_parentales_OR']]
```

```python
corrOR[(corrOR['PVal_todos_OR']<0.05) | (corrOR['PVal_parentales_OR']<0.05)]#todos los signifivativos OR
```

```python
#entonces voy a crear df por separado para todos los sign en CE y en OR
```

```python
sign_CE = corrCE[(corrCE['PVal_todos_CE']<0.05) | (corrCE['PVal_parentales_CE']<0.05)]#todos los signifivativos CE

sign_OR = corrOR[(corrOR['PVal_todos_OR']<0.05) | (corrOR['PVal_parentales_OR']<0.05)]#todos los signifivativos OR

print('dim sign_CE: ' , sign_CE.shape , 'dim sign_OR: ' , sign_OR.shape)
```

```python
#puedo buscar en cada df a los que son sign exclusivamente en Todos, a los q son sign exclusivamente en parentales
#a los que son sign en ambos grupos de genotipos simultaneamente. Para cada estadio celular por separado.

#luego puedo hacer un merge entre ambos df, y ver de ese merge las combinaciones posibles
#es decir, evaluar primero si hay loci en comun entre ambos estadios celulares, que significativamente correlacione
#transcripto con proteina, y en caso afirmativo, evaluar si es por parentales, o incluyendo hibridos
```

```python
#arrancamos con CE, los sign exclusivamente parentales
parentales_significativos_CE = sign_CE[(sign_CE['PVal_parentales_CE'] < 0.5) & (sign_CE['PVal_todos_CE'] > 0.5)]
```

```python
#arrancamos con CE, los sign exclusivamente todos(con hibridos incluidos)
sign_CE[(sign_CE['PVal_parentales_CE'] > 0.5) & (sign_CE['PVal_todos_CE'] < 0.5)]
```

```python
#arrancamos con CE, los sign simultaneamente todos(con hibridos incluidos) y parentales
todos_parentales_significativos_CE = sign_CE[(sign_CE['PVal_parentales_CE'] < 0.5) & (sign_CE['PVal_todos_CE'] < 0.5)]
```

```python
#arrancamos con CE, los NOsign simultaneamente todos(con hibridos incluidos) y parentales
sign_CE[(sign_CE['PVal_parentales_CE'] > 0.5) & (sign_CE['PVal_todos_CE'] > 0.5)]
```

```python
####################
print('sign_CE: ',sign_CE.shape[0] , 'parentales_sign_CE: ',parentales_significativos_CE.shape[0] ,
      'sign_todos_par :',todos_parentales_significativos_CE.shape[0])
```

```python
#de aca se desprende que hay 5 casos en los que entre parentales son sign, y al incluir el hib, dejan de serlo
#y que hay 165 casos en los que es sign incluyendo o no al hib
```

```python
#ahora hago le mismo analisis para OR
#arrancamos con OR, los sign exclusivamente parentales
parentales_significativos_OR = sign_OR[(sign_OR['PVal_parentales_OR'] < 0.5) & (sign_OR['PVal_todos_OR'] > 0.5)]
```

```python
#arrancamos con OR, los sign exclusivamente todos(par+hib)
todos_significativos_OR = sign_OR[(sign_OR['PVal_parentales_OR'] > 0.5) & (sign_OR['PVal_todos_OR'] < 0.5)]
```

```python
#arrancamos con OR, los sign simultaneamente [todos(par+hib)] + parentales(solos)
todos_parentales_significativos_or = sign_OR[(sign_OR['PVal_parentales_OR'] < 0.5) & (sign_OR['PVal_todos_OR'] < 0.5)]
```

```python
#arrancamos con OR, los NOsign simultaneamente [todos(par+hib)] + parentales(solos)
sign_OR[(sign_OR['PVal_parentales_OR'] > 0.5) & (sign_OR['PVal_todos_OR'] > 0.5)]
```

```python
print('sign_OR: ' ,sign_OR.shape[0] ,'sign_parentales_OR: ' ,parentales_significativos_OR.shape[0] ,
'todos_sign_OR: ',todos_significativos_OR.shape[0] ,
'todos_parentales_sign_OR:' ,todos_parentales_significativos_or.shape[0])
```

```python
#de aca se desprende que solo entre parentales hay 3 loci que son sign, y dejan de serlo cuando se incluye al hib
#que hay 1 caso donde al incluir al h'ibrido comienza a ser significativo, pero entre los parentales, no lo era
#155 casos donde simultaneamente correlacionan significativamente transcripto y proteina 
```
