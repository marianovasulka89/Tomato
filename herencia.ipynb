```python
#additive (A) and dominance (D)
#[A = (P1-P2)/2]#aca es re importante el orden p1 p2 xq resta
#[D=H-((P1+P2)/2)]#aca el orden p1 p2 no importa, xq suman
#(D/A)
#(OR;D/A < 1.2), recessive (R; -1.2 ≤ D/A ≤ -0.8), additive (A; -0.8 < D/A < 0.8), 
#dominant (D; 0.8 ≤ D/A ≤ 1.2), or overdominant (OD; D/A > 1.2)

#con lycopersicum como p1 y ceraciforme como p2

#o sea, levovil - cervil
#LA0147 - Plovdiv
#Ferum - LA1420

for fila in range(0,p.shape[0]):
    #para prot_CE
    
    A_LxC_P_CE = ((float(p.loc[fila , ['LevovilCEMean']]) - float(p.loc[fila , ['CervilCEMean']]))/2)
    
    D_LxC_P_CE = ((float(p.loc[fila , ['LevovilxCervilCEMean']])) - ((float(p.loc[fila , ['CervilCEMean']]) + float(p.loc[fila , ['LevovilCEMean']]))/2))
    
    A_FxLA1420_P_CE = ((float(p.loc[fila , ['FerumCEMean']]) - float(p.loc[fila , ['LA1420CEMean']]))/2)
    
    D_FxLA1420_P_CE = ((float(p.loc[fila , ['FerumxLA1420CEMean']])) - ((float(p.loc[fila , ['LevovilCEMean']]) + float(p.loc[fila , ['CervilCEMean']]))/2))
    
    A_PxLA0147_P_CE = ((float(p.loc[fila , ['LA0147CEMean']]) - float(p.loc[fila , ['PlovdivCEMean']]))/2)
    
    D_PxLA0147_P_CE = ((float(p.loc[fila , ['LA0147xPlovdivCEMean']])) - ((float(p.loc[fila , ['PlovdivCEMean']]) + float(p.loc[fila , ['LA0147CEMean']]))/2))
    
    #para prot_OR
    
    A_LxC_P_OR = ((float(p.loc[fila , ['LevovilORMean']]) - float(p.loc[fila , ['CervilORMean']]))/2)
    
    D_LxC_P_OR = ((float(p.loc[fila , ['LevovilxCervilORMean']])) - ((float(p.loc[fila , ['CervilORMean']]) + float(p.loc[fila , ['LevovilORMean']]))/2))
    
    A_FxLA1420_P_OR = ((float(p.loc[fila , ['FerumORMean']]) - float(p.loc[fila , ['LA1420ORMean']]))/2)
    
    D_FxLA1420_P_OR = ((float(p.loc[fila , ['FerumxLA1420ORMean']])) - ((float(p.loc[fila , ['FerumORMean']]) + float(p.loc[fila , ['LA1420ORMean']]))/2))
    
    A_PxLA0147_P_OR = ((float(p.loc[fila , ['LA0147ORMean']]) - float(p.loc[fila , ['PlovdivORMean']]))/2)
    
    D_PxLA0147_P_OR = ((float(p.loc[fila , ['LA0147xPlovdivORMean']])) - ((float(p.loc[fila , ['PlovdivORMean']]) + float(p.loc[fila , ['LA0147ORMean']]))/2))
    
    #para trans
    
    A_LxC_T = ((float(t.loc[fila , ['LevovilMean']]) - float(t.loc[fila , ['CervilMean']]))/2)
    
    D_LxC_T = ((float(t.loc[fila , ['Levovil_x_CervilMean']])) - ((float(t.loc[fila , ['CervilMean']]) + float(t.loc[fila , ['LevovilMean']]))/2))
    
    A_FxLA01420_T = ((float(t.loc[fila , ['FerumMean']]) - float(t.loc[fila , ['LA1420Mean']]))/2)
    
    D_FxLA01420_T = ((float(t.loc[fila , ['LA1420_x_FerumMean']])) - ((float(t.loc[fila , ['FerumMean']]) + float(t.loc[fila , ['LA1420Mean']]))/2))
    
    A_PxLA0147_T = ((float(t.loc[fila , ['LA0147Mean']]) - float(t.loc[fila , ['PlovdivMean']]))/2)
    
    D_PxLA0147_T = ((float(t.loc[fila , ['LA0147_x_PlovdivMean']])) - ((float(t.loc[fila , ['PlovdivMean']]) + float(t.loc[fila , ['LA0147Mean']]))/2))
    
    #ahora queda poner en cada df_P_CE df_P_OR o df_T los resultados calculados, en la fila correspondiente
    
    #3A
    
    df_P_CE.loc[fila , ['A_LxC']] = A_LxC_P_CE
    
    df_P_CE.loc[fila , ['A_FxLA1420']] = A_FxLA1420_P_CE
    
    df_P_CE.loc[fila , ['A_PxLA0147']] = A_PxLA0147_P_CE
    
    #3D
    
    df_P_CE.loc[fila , ['D_LxC']] = D_LxC_P_CE
    
    df_P_CE.loc[fila , ['D_FxLA1420']] = D_FxLA1420_P_CE
    
    df_P_CE.loc[fila , ['D_PxLA0147']] = D_PxLA0147_P_CE
    
    #3A
    
    df_P_OR.loc[fila , ['A_LxC']] = A_LxC_P_OR
    
    df_P_OR.loc[fila , ['A_FxLA1420']] = A_FxLA1420_P_OR
    
    df_P_OR.loc[fila , ['A_PxLA0147']] = A_PxLA0147_P_OR
    
    #3D
    
    df_P_OR.loc[fila , ['D_LxC']] = D_LxC_P_OR
    
    df_P_OR.loc[fila , ['D_FxLA1420']] = D_FxLA1420_P_OR
    
    df_P_OR.loc[fila , ['D_PxLA0147']] = D_PxLA0147_P_OR
    
    #3A
    
    df_T.loc[fila , ['A_LxC']] = A_LxC_T
    
    df_T.loc[fila , ['A_FxLA1420']] = A_FxLA01420_T
    
    df_T.loc[fila , ['A_PxLA0147']] = A_PxLA0147_T
    
    #3D
    
    df_T.loc[fila , ['D_LxC']] = D_LxC_T
    
    df_T.loc[fila , ['D_FxLA1420']] = D_FxLA01420_T
    
    df_T.loc[fila , ['D_PxLA0147']] = D_PxLA0147_T
```
